```python
def get_all_line_url():
    res = requests.get(URL)
    data = res.content.decode('gbk')
    soup = BeautifulSoup(data,"html.parser")
    lines = soup.find_all(name='ul',attrs={"class":"zhan_shike"})
    all_lines = {}
    for i in lines[0].find_all(name='span'):
        all_lines.update({
            re.findall(r"""/">(.*?)</a>""", str(i))[0]:
            re.findall(r"""href="(.*?)">""", str(i))[0]
        })
    return all_lines
    
all_lines = get_all_line_url()
```

```python
def get_line_station(url, key):
#     url = 'http://ditie.114huoche8.com/BeiJing/x_1/'
    headers={'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'}
    res = requests.get(url, headers=headers)
    data = res.content.decode('gbk')
    soup = BeautifulSoup(data,"html.parser")
    stations = soup.find_all(name='div',attrs={"class":"table"})
    stations = stations[0].find_all(name='li',attrs={"class":None})
    d = []
    for i in stations:
        d.append({
            'station': i.a.string,
            'time': i.find_all('span')[2].string,
        })
    df = pd.DataFrame(d)
    df = df.set_index('station')
    print(df)
    df.to_csv('{}.csv'.format(key.replace('/', '&')))
```

```python
subway = {} # 保存站点归属、时间信息
all_relations = {} # 保存站点联通关系
print(len(all_lines))
for i in all_lines:
    df = pd.read_csv('{}.csv'.format(i.replace('/', '&')))
    l = df['station'].tolist()
    df = df.set_index('station')
    for j, k in enumerate(l):
        i = i.replace('/', '&')
        if j == 0 and i in ['北京地铁2号线', '北京地铁10号线']:
            subway['-'.join([l[-1], l[j]])] = [i, 3]
            subway['-'.join([l[j], l[-1]])] = [i, 3]
            
        if j == (len(l) - 1):
            pass
        else:
            if df.loc[l[j], 'time'] in ['—', '-', 'nan'] or  df.loc[l[j + 1], 'time'] in ['—', '-', 'nan'] \
            or pd.isnull(df.loc[l[j], 'time']) or pd.isnull(df.loc[l[j + 1], 'time']):
                subway['-'.join([l[j], l[j + 1]])] = [i, 3]
                subway['-'.join([l[j + 1], l[j]])] = [i, 3]
            else:
                time = (datetime.strptime(df.loc[l[j + 1], 'time'], '%H:%M') - datetime.strptime(df.loc[l[j], 'time'], '%H:%M')).seconds
                time = time / 60 if time > 0 else (0 - time) /60
                subway['-'.join([l[j], l[j + 1]])] = [i, time]
                subway['-'.join([l[j + 1], l[j]])] = [i, time]
        
        
        if not all_relations.get(k):
            all_relations[k] = []
        if j == 0 and i in ['北京地铁2号线', '北京地铁10号线']:
            all_relations[k].append(l[-1])
        elif j == 0:
            pass
        else:
            all_relations[k].append(l[j - 1])
            
        if j == (len(l) -1 ) and i in ['北京地铁2号线', '北京地铁10号线']:
            all_relations[k].append(l[0])
        elif j == (len(l) -1 ):
            pass
        else:
            all_relations[k].append(l[j + 1])
        
```

```python
def search(start, destination, connection_grpah, sort_candidate):
    pathes = [[start]]
    
#     visitied = set()
    while pathes: 

        path = pathes.pop(0)
        froninter = path[-1]
        if destination in path:
            pathes.append(path)
            new_pathes, new_nums, new_func = sort_candidate(pathes)
            if new_func(path) == new_nums: 
                return path, new_nums, new_func
        else:
        
#         if froninter in visitied: continue 这里错误 如果环存在怎么解决？
            
            successors = connection_grpah[froninter]

            for station in successors:
                if station in path: continue  

                new_path = path + [station]

                pathes.append(new_path)
                new_pathes, new_nums, new_func = sort_candidate(pathes)
                if new_func(new_path) != new_nums: # 根据排序方式判断 时间复杂度问题？
                    continue
                if station == destination: 
                    return new_path, new_nums, new_func
        
        pathes, num, func = sort_candidate(pathes)
        
def transfer_stations_first(pathes): 
    return sorted(pathes, key=len)
```

```python
def shortest_time_first(pathes):
    
    def get_sum_times(path):
        sum_times = 0
        for i, j in enumerate(path):
            if i < (len(path) -1):
                sum_times += int(subway['-'.join([path[i], path[i + 1]])][1])
        return sum_times
    
    if len(pathes) <= 1: return pathes, 0, get_sum_times
   
    ret = sorted(pathes, key=get_sum_times)
    return ret, get_sum_times(ret[0]), get_sum_times
```

```python
def shortest_change_first(pathes):
    
    def get_change_times(path):
        changes = []
        for i, j in enumerate(path):
            if i < (len(path) - 1):
                changes.append(subway['-'.join([path[i], path[i + 1]])][0])
        return len(set(changes))
    if len(pathes) <= 1: return pathes, 0, get_change_times
    
    ret = sorted(pathes, key=get_change_times)
    return ret, get_change_times(ret[0]), get_change_times
```

#### 解决了最后一次如果是换乘，返回的最少换乘不对的问题

```python
# 还是有小问题，比如最短换乘存在两个的问题
```

```python
def the_composite(start, destination, connection_grpah):
    
    shortest_time, time, time_func = search(start, destination, all_relations, shortest_time_first)
    shortest_change, change, change_func = search(start, destination, all_relations, shortest_change_first)
    df = pd.DataFrame([
        {
            'strategy': shortest_time,
            'time': time,
            'change': change_func(shortest_time),
        },
        {
            'strategy': shortest_change,
            'time': time_func(shortest_time),
            'change': change,
        },
    ])
    df['change_rank'] = df['change'].rank(method='min')
    df['time_rank'] = df['time'].rank(method='min')
    df['temp'] = df['change_rank'] + df['time_rank']
    df['rank'] = df['temp'].rank()
    df = df.sort_values('rank')
    df = df[[
        'strategy',
        'change',
        'time',
    ]]
    return df.iloc[0]
    
```

```python
the_composite('三元桥', '五道口', all_relations)
```

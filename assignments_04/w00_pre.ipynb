```python
from hanziconv import HanziConv
import re
import sys
```

```python
input_file = '../../data/wiki_00'
p1 = re.compile('（）')
p2 = re.compile('《》')
p3 = re.compile('「')
p4 = re.compile('」')
p5 = re.compile('<doc (.*)>')
p6 = re.compile('</doc>')
outfile = open(input_file + 'simple', 'w')
with open(input_file, 'r') as f:
    lines = f.readlines()
    for i, line in enumerate(lines):
        if i % 10000 == 0:
            print(i)
        line = p1.sub('', line)
        line = p2.sub('', line)
        line = p3.sub('', line)
        line = p4.sub('', line)
        line = p5.sub('', line)
        line = p6.sub('', line)
        line = HanziConv.toSimplified(line)
        outfile.write(line)
outfile.close()
```
